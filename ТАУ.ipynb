1. Критерий управляемости и наблюдаемости


```python
import numpy as np

print("Заполните матрицу А построчно: ")
rows = int(input("Введите количество строк: "))
cols = int(input("Введите количество столбцов: "))
A = np.array([list(map(int, input(f"Строка {i+1}: ").split()))
                  for i in range(rows)])
print("Введите вектор-столбец B:")
B = np.array([list(map(int, input(f"Строка {i+1}: ").split()))
                  for i in range(1)]).reshape(-1,1)
print("Введите строку С:")
C = np.array([list(map(int, input(f"Строка {i+1}: ").split()))
                  for i in range(1)])
B1 = B.copy()
C1 = C.copy()
Su = np.zeros((len(A[0]),len(A[0])),dtype=int) #Матрица управления
Su[:,0] = B.reshape(3)
Sn = np.zeros((len(A[0]),len(A[0])),dtype=int) #Матрица наблюдения
Sn[0,:] = C

for i in range(1,len(A[0])):
  Su[:,i] = np.dot(A,B1.reshape(3))
  B1 = np.dot(A,B1)
for i in range(1,len(A[0])):
  Sn[i,:] = np.dot(C1,A)
  C1 = np.dot(C1,A)
print(f"Матрица управления\n{Su},\nМатрица наблюдения\n{Sn}")

ru = np.linalg.matrix_rank(Su)
print(f"Ранг матрицы управления:{ru}")
if 0< np.linalg.matrix_rank(Su)< len(A[0]):
  print("Система частично управляема")
elif np.linalg.matrix_rank(Su) == len(A[0]):
  print("Система полностью управляема")
else:
  print("Система неуправялема")

rn = np.linalg.matrix_rank(Sn)
print(f"Ранг матрицы наблюдения:{rn}")
if 0< np.linalg.matrix_rank(Sn)< len(A[0]):
  print("Система частично наблюдаема")
elif np.linalg.matrix_rank(Sn) == len(A[0]):
  print("Система полностью наблюдаема")
else:
  print("Система ненаблюдаема")
```

2. Приведение к диагональному виду


```python

print("Заполните матрицу А построчно: ")
rows = int(input("Введите количество строк: "))
cols = int(input("Введите количество столбцов: "))
A = np.array([list(map(int, input(f"Строка {i+1}: ").split()))
                  for i in range(rows)])
print("Введите вектор-столбец B:")
B = np.array([list(map(int, input(f"Строка {i+1}: ").split()))
                  for i in range(1)]).reshape(-1,1)
print("Введите строку С:")
C = np.array([list(map(int, input(f"Строка {i+1}: ").split()))
                  for i in range(1)])
eigenvalues, eigenvectors = np.linalg.eig(A)

sorted_indices = np.argsort(eigenvalues)
L = np.diag(eigenvalues[sorted_indices]).astype(complex)
R = eigenvectors[:, sorted_indices]
print(L,'\n',R)
Bd = np.dot(np.linalg.inv(R),B) # управляемость
Cd = np.dot(C,R) # наблюдаемость
print(f"Матрица В диагональная \n{Bd} \nМатрица С диагональная \n{Cd}")

tolerance = 1e-10
for i in range(0,len(Bd)):
  if np.abs(Bd[i,0]) < tolerance:
    print(f"Система не управляема по {i+1} координате")
  else:
    print(f"Система управляема по {i+1} координате")
for i in range(0,len(Cd[0])):
  if np.abs(Cd[0,i]) < tolerance:
    print(f"Система не наблюдаема по {i+1} координате")
  else:
    print(f"Система наблюдаема по {i+1} координате")
```
